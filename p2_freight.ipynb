```python
# === 2) 번들에서 복원 ===
import pickle

BUNDLE_PATH = "_snapshots/olist_clean_bundle.pkl"  # 경로만 맞추면 됩니다.

with open(BUNDLE_PATH, "rb") as f:
    bundle = pickle.load(f)

# 변수로 바로 풀어서 사용
orders_clean       = bundle["orders_clean"]
order_items_clean  = bundle["order_items_clean"]
payments_clean     = bundle["payments_clean"]
reviews_clean      = bundle["reviews_clean"]
customers_clean    = bundle["customers_clean"]
geolocation_clean  = bundle["geolocation_clean"]
products_clean     = bundle["products_clean"]
sellers_clean      = bundle["sellers_clean"]
category_tr_clean  = bundle["category_tr_clean"]

print("✅ Restored all *_clean DataFrames")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 160)

# =========================================================
# 과제 2: 주문→배송 리드타임 분석 + 지역별 성과 + 리뷰 영향
# 입력 DF: orders_clean, customers_clean, reviews_clean
# =========================================================

# ---------- 1) 배송완료 주문에서 리드타임/지연 계산 ----------
use_cols = [
    "order_id","customer_id",
    "order_purchase_timestamp",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "order_status"
]
o = orders_clean[use_cols].copy()
o = o[o["order_status"]=="delivered"].dropna(subset=[
    "order_purchase_timestamp","order_delivered_customer_date","order_estimated_delivery_date"
])

# 날짜 차이(일)
o["proc_days"]   = (o["order_delivered_carrier_date"]  - o["order_purchase_timestamp"]).dt.total_seconds()/86400
o["ship_days"]   = (o["order_delivered_customer_date"] - o["order_delivered_carrier_date"]).dt.total_seconds()/86400
o["lead_days"]   = (o["order_delivered_customer_date"] - o["order_purchase_timestamp"]).dt.total_seconds()/86400
o["delay_signed"]= (o["order_delivered_customer_date"] - o["order_estimated_delivery_date"]).dt.total_seconds()/86400  # 음수=조기
o["delay_days"]  = o["delay_signed"].clip(lower=0)  # 지연일수(조기는 0)
o["is_late"]     = o["delay_days"]>0

# 고객 주(state) 조인
cust_state = customers_clean[["customer_id","customer_unique_id","customer_state"]].drop_duplicates("customer_id")
o = o.merge(cust_state, on="customer_id", how="left")

# 리뷰(주문단위 평균점수) 조인
if "review_score" in reviews_clean.columns:
    rev = (reviews_clean.assign(review_score=pd.to_numeric(reviews_clean["review_score"], errors="coerce"))
                     .groupby("order_id", as_index=False)["review_score"].mean())
    o = o.merge(rev, on="order_id", how="left")
else:
    o["review_score"] = pd.NA

# ---------- 2) 지역별 KPI ----------
def _nz(x): 
    return float(x) if pd.notnull(x) else 0.0

state_kpi = (o.groupby("customer_state")
               .agg(
                   orders=("order_id","nunique"),
                   lead_mean=("lead_days","mean"),
                   lead_median=("lead_days","median"),
                   lead_p90=("lead_days", lambda s: s.quantile(0.90)),
                   late_rate=("is_late","mean"),
                   late_days_mean=("delay_days", lambda s: s[s>0].mean()),
                   review_mean=("review_score","mean")
               )
               .reset_index())

# 보기 좋게 가공
state_kpi["late_rate(%)"] = (state_kpi["late_rate"]*100).round(2)
for c in ["lead_mean","lead_median","lead_p90","late_days_mean","review_mean"]:
    state_kpi[c] = state_kpi[c].round(2)

# TOP 리스트(지연율, 리드타임, 물량)
top_late = state_kpi.sort_values("late_rate(%)", ascending=False).head(10)
top_lead = state_kpi.sort_values("lead_mean", ascending=False).head(10)
top_vol  = state_kpi.sort_values("orders", ascending=False).head(10)

print("=== [주별 KPI] ===")
display(state_kpi.sort_values("orders", ascending=False).head(20))
print("\n[지연율 TOP 10]");   display(top_late[["customer_state","orders","late_rate(%)","lead_mean","review_mean"]])
print("\n[리드타임 TOP 10]"); display(top_lead[["customer_state","orders","lead_mean","lead_p90","late_rate(%)"]])
print("\n[물량 TOP 10]");     display(top_vol [["customer_state","orders","lead_mean","late_rate(%)"]])

# ---------- 3) 월별 추이(리드타임, 지연율) ----------
o["order_month"] = o["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()
monthly_kpi = (o.groupby("order_month")
                 .agg(orders=("order_id","nunique"),
                      lead_mean=("lead_days","mean"),
                      late_rate=("is_late","mean"))
                 .reset_index())
monthly_kpi["late_rate(%)"] = (monthly_kpi["late_rate"]*100).round(2)
monthly_kpi["lead_mean"]    = monthly_kpi["lead_mean"].round(2)

print("\n=== [월별 추이] ===")
display(monthly_kpi.head(12))

# ---------- 4) 지연-리뷰 영향(바이너리/구간) ----------
# 지연 카테고리
def _delay_bucket(x):
    if x<=0:     return "정시/조기"
    if x<=3:     return "1-3일 지연"
    if x<=7:     return "4-7일 지연"
    return ">7일 지연"
o["delay_bucket"] = o["delay_signed"].apply(_delay_bucket)

# 저평점(≤3) 비율과 평균점수
delay_bins_kpi = (o.groupby("delay_bucket")
                    .agg(orders=("order_id","nunique"),
                         review_mean=("review_score","mean"),
                         low_rate=("review_score", lambda s: (s<=3).mean()))
                    .reset_index())
delay_bins_kpi["low_rate(%)"] = (delay_bins_kpi["low_rate"]*100).round(2)
delay_bins_kpi["review_mean"] = delay_bins_kpi["review_mean"].round(2)
print("\n=== [지연구간별 리뷰] ===")
display(delay_bins_kpi)

# ---------- 5) 시각화 ----------
# 팔레트(알록달록)
colors = ["#4E79A7","#F28E2B","#E15759","#76B7B2","#59A14F",
          "#EDC948","#B07AA1","#FF9DA7","#9C755F","#2E7C9E"]

# (A) 총 리드타임 분포 히스토그램
plt.figure(figsize=(8,5))
vals = o["lead_days"].dropna()
plt.hist(vals, bins=40, color=colors[0], alpha=0.85, edgecolor="white")
med, p90 = vals.median(), vals.quantile(0.90)
plt.axvline(med, color=colors[2], linestyle="--", linewidth=2, label=f"Median={med:.1f}d")
plt.axvline(p90, color=colors[1], linestyle="--", linewidth=2, label=f"P90={p90:.1f}d")
plt.title("총 리드타임 분포(일)")
plt.xlabel("lead_days"); plt.ylabel("주문수")
plt.legend(); plt.tight_layout(); plt.show()

# (B) 주별 지연율 TOP 12 바
top12 = state_kpi.sort_values("late_rate(%)", ascending=False).head(12)
plt.figure(figsize=(10,5))
plt.bar(top12["customer_state"], top12["late_rate(%)"], color=colors[:len(top12)])
plt.title("주별 지연율 TOP12(%)"); plt.xlabel("State"); plt.ylabel("지연율(%)")
plt.xticks(rotation=30); plt.tight_layout(); plt.show()

# (C) 주별 버블맵: x=중앙 리드타임, y=지연율, s=물량
plt.figure(figsize=(9,6))
sx = state_kpi["lead_median"].values
sy = state_kpi["late_rate(%)"].values
ss = (state_kpi["orders"]/state_kpi["orders"].max()*2200).clip(60, 2200).values
for i, r in state_kpi.reset_index(drop=True).iterrows():
    c = colors[i % len(colors)]
    plt.scatter(r["lead_median"], r["late_rate(%)"], s=ss[i], color=c, alpha=0.9, edgecolor="white", linewidth=0.7)
    plt.text(r["lead_median"], r["late_rate(%)"], r["customer_state"], ha="center", va="center", fontsize=9, color="white",
             bbox=dict(boxstyle="round,pad=0.2", fc=c, ec="none", alpha=0.9))
plt.title("주별 성과 버블맵: 중앙 리드타임 vs 지연율 (크기=물량)")
plt.xlabel("lead_median(일)"); plt.ylabel("지연율(%)")
plt.tight_layout(); plt.show()

# (D) 월별 추이: 리드타임(막대) + 지연율(선)
plt.figure(figsize=(10,5))
ax = plt.gca()
ax.bar(monthly_kpi["order_month"], monthly_kpi["lead_mean"], color=colors[3], width=20)
ax.set_ylabel("평균 리드타임(일)")
ax2 = ax.twinx()
ax2.plot(monthly_kpi["order_month"], monthly_kpi["late_rate(%)"], color=colors[2], linewidth=2, marker="o")
ax2.set_ylabel("지연율(%)")
plt.title("월별 평균 리드타임 & 지연율")
plt.tight_layout(); plt.show()

# (E) 지연구간별 리뷰: 막대(저평점 비율) + 선(평균점수)
x = delay_bins_kpi["delay_bucket"]
plt.figure(figsize=(8,5))
ax = plt.gca()
ax.bar(x, delay_bins_kpi["low_rate(%)"], color=colors[:len(x)])
ax.set_ylabel("저평점 비율(≤3) [%]")
ax2 = ax.twinx()
ax2.plot(x, delay_bins_kpi["review_mean"], color="#333333", marker="o", linewidth=2)
ax2.set_ylabel("평균 리뷰 점수")
plt.title("지연구간별 리뷰 반응")
plt.tight_layout(); plt.show()

# ---------- 6) 간단 결론(프린트) ----------
def _line(k,v): print(f"- {k}: {v}")

print("\n=== 핵심 요약 ===")
_line("전체 평균 리드타임(일)", round(o["lead_days"].mean(),2))
_line("P90(일)", round(o["lead_days"].quantile(0.90),2))
_line("전체 지연율(%)", round(o["is_late"].mean()*100,2))
if pd.api.types.is_numeric_dtype(o["review_score"]):
    # 상관 근사(선형상관, 참고용)
    corr = pd.Series(o["delay_signed"]).corr(pd.Series(o["review_score"]))
    _line("지연(서명)↔리뷰 상관계수", round(corr,3))
_line("지연율 높은 주 TOP3", ", ".join(top_late.head(3)["customer_state"].tolist()))
_line("물량 많은 주 TOP3", ", ".join(top_vol.head(3)["customer_state"].tolist()))
```

```python
# =========================================================
# 과제 2: 주문→배송 리드타임 분석 + 지역별 성과 + 리뷰 영향
# 요구사항 반영:
# - [월별 평균 리드타임 & 지연율] 그래프 제목에 "(막대=평균 리드타임, 선=지연율)" 명시
# - [지연구간별 리뷰 반응] 그래프 제목에 "(막대=저평점 비율, 선=평균 리뷰 점수)" 명시
# 전제: orders_clean, customers_clean, reviews_clean 이 메모리에 존재
# (선행 셀에서 matplotlib.pyplot as plt, numpy as np, pandas as pd 이미 import됨)
# =========================================================

# ---------- 1) 배송완료 주문에서 리드타임/지연 계산 ----------
use_cols = [
    "order_id","customer_id",
    "order_purchase_timestamp",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "order_status"
]
o = orders_clean[use_cols].copy()
o = o[o["order_status"]=="delivered"].dropna(subset=[
    "order_purchase_timestamp","order_delivered_customer_date","order_estimated_delivery_date"
])

# 날짜 차이(일)
o["proc_days"]    = (o["order_delivered_carrier_date"]  - o["order_purchase_timestamp"]).dt.total_seconds()/86400
o["ship_days"]    = (o["order_delivered_customer_date"] - o["order_delivered_carrier_date"]).dt.total_seconds()/86400
o["lead_days"]    = (o["order_delivered_customer_date"] - o["order_purchase_timestamp"]).dt.total_seconds()/86400
o["delay_signed"] = (o["order_delivered_customer_date"] - o["order_estimated_delivery_date"]).dt.total_seconds()/86400  # 음수=조기
o["delay_days"]   = o["delay_signed"].clip(lower=0)  # 지연일수(조기는 0)
o["is_late"]      = o["delay_days"]>0

# 고객 주(state) 조인
cust_state = customers_clean[["customer_id","customer_unique_id","customer_state"]].drop_duplicates("customer_id")
o = o.merge(cust_state, on="customer_id", how="left")

# 리뷰(주문단위 평균점수) 조인
if "review_score" in reviews_clean.columns:
    rev = (reviews_clean.assign(review_score=pd.to_numeric(reviews_clean["review_score"], errors="coerce"))
                     .groupby("order_id", as_index=False)["review_score"].mean())
    o = o.merge(rev, on="order_id", how="left")
else:
    o["review_score"] = pd.NA

# ---------- 2) 지역별 KPI ----------
state_kpi = (o.groupby("customer_state")
               .agg(
                   orders=("order_id","nunique"),
                   lead_mean=("lead_days","mean"),
                   lead_median=("lead_days","median"),
                   lead_p90=("lead_days", lambda s: s.quantile(0.90)),
                   late_rate=("is_late","mean"),
                   late_days_mean=("delay_days", lambda s: s[s>0].mean()),
                   review_mean=("review_score","mean")
               )
               .reset_index())

state_kpi["late_rate(%)"] = (state_kpi["late_rate"]*100).round(2)
for c in ["lead_mean","lead_median","lead_p90","late_days_mean","review_mean"]:
    state_kpi[c] = state_kpi[c].round(2)

# TOP 리스트
top_late = state_kpi.sort_values("late_rate(%)", ascending=False).head(10)
top_lead = state_kpi.sort_values("lead_mean",    ascending=False).head(10)
top_vol  = state_kpi.sort_values("orders",       ascending=False).tail(10) if hasattr(state_kpi["orders"], "tail") else state_kpi.sort_values("orders", ascending=False).head(10)
# (일부 환경에서 descending 옵션이 없을 수 있어 보수적으로 처리)

print("=== [주별 KPI] ===")
display(state_kpi.sort_values("orders", ascending=False).head(20))
print("\n[지연율 TOP 10]");   display(top_late[["customer_state","orders","late_rate(%)","lead_mean","review_mean"]])
print("\n[리드타임 TOP 10]"); display(top_lead[["customer_state","orders","lead_mean","lead_p90","late_rate(%)"]])
print("\n[물량 TOP 10]");     display(state_kpi.sort_values("orders", ascending=False).head(10)[["customer_state","orders","lead_mean","late_rate(%)"]])

# ---------- 3) 월별 추이(리드타임, 지연율) ----------
o["order_month"] = o["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()
monthly_kpi = (o.groupby("order_month")
                 .agg(orders=("order_id","nunique"),
                      lead_mean=("lead_days","mean"),
                      late_rate=("is_late","mean"))
                 .reset_index())
monthly_kpi["late_rate(%)"] = (monthly_kpi["late_rate"]*100).round(2)
monthly_kpi["lead_mean"]    = monthly_kpi["lead_mean"].round(2)

print("\n=== [월별 추이] ===")
display(monthly_kpi.head(12))

# ---------- 4) 지연-리뷰 영향(바이너리/구간) ----------
def _delay_bucket(x):
    if x<=0:     return "정시/조기"
    if x<=3:     return "1-3일 지연"
    if x<=7:     return "4-7일 지연"
    return ">7일 지연"
o["delay_bucket"] = o["delay_signed"].apply(_delay_bucket)

delay_bins_kpi = (o.groupby("delay_bucket")
                    .agg(orders=("order_id","nunique"),
                         review_mean=("review_score","mean"),
                         low_rate=("review_score", lambda s: (s<=3).mean()))
                    .reset_index())
delay_bins_kpi["low_rate(%)"]  = (delay_bins_kpi["low_rate"]*100).round(2)
delay_bins_kpi["review_mean"]  = delay_bins_kpi["review_mean"].round(2)

print("\n=== [지연구간별 리뷰] ===")
display(delay_bins_kpi)

# ---------- 5) 시각화 ----------
# 팔레트(알록달록)
colors = ["#4E79A7","#F28E2B","#E15759","#76B7B2","#59A14F",
          "#EDC948","#B07AA1","#FF9DA7","#9C755F","#2E7C9E"]

# (A) 총 리드타임 분포 히스토그램
plt.figure(figsize=(8,5))
vals = o["lead_days"].dropna()
bars = plt.hist(vals, bins=40, color=colors[0], alpha=0.85, edgecolor="white")
med, p90 = vals.median(), vals.quantile(0.90)
l1 = plt.axvline(med, color=colors[2], linestyle="--", linewidth=2, label=f"Median={med:.1f}d")
l2 = plt.axvline(p90, color=colors[1], linestyle="--", linewidth=2, label=f"P90={p90:.1f}d")
plt.title("총 리드타임 분포(일)")
plt.xlabel("lead_days"); plt.ylabel("주문수")
plt.legend(); plt.tight_layout(); plt.show()

# (B) 주별 지연율 TOP 12 바
top12 = state_kpi.sort_values("late_rate(%)", ascending=False).head(12)
plt.figure(figsize=(10,5))
plt.bar(top12["customer_state"], top12["late_rate(%)"], color=colors[:len(top12)], label="지연율(%)")
plt.title("주별 지연율 TOP12(%) — (막대=지연율)")
plt.xlabel("State"); plt.ylabel("지연율(%)")
plt.xticks(rotation=30); plt.legend()
plt.tight_layout(); plt.show()

# (C) 주별 버블맵: x=중앙 리드타임, y=지연율, s=물량
plt.figure(figsize=(9,6))
sx = state_kpi["lead_median"].values
sy = state_kpi["late_rate(%)"].values
ss = (state_kpi["orders"]/state_kpi["orders"].max()*2200).clip(60, 2200).values
for i, r in state_kpi.reset_index(drop=True).iterrows():
    c = colors[i % len(colors)]
    plt.scatter(r["lead_median"], r["late_rate(%)"], s=ss[i], color=c, alpha=0.9, edgecolor="white", linewidth=0.7)
    plt.text(r["lead_median"], r["late_rate(%)"], r["customer_state"], ha="center", va="center", fontsize=9, color="white",
             bbox=dict(boxstyle="round,pad=0.2", fc=c, ec="none", alpha=0.9))
plt.title("주별 성과 버블맵: 중앙 리드타임 vs 지연율 (크기=물량)")
plt.xlabel("lead_median(일)"); plt.ylabel("지연율(%)")
plt.tight_layout(); plt.show()

# (D) 월별 추이: 리드타임(막대) + 지연율(선) — 의미를 제목에 명시
plt.figure(figsize=(10,5))
ax = plt.gca()
bar = ax.bar(monthly_kpi["order_month"], monthly_kpi["lead_mean"], color=colors[3], width=20, label="평균 리드타임(일)")
ax.set_ylabel("평균 리드타임(일)")
ax2 = ax.twinx()
line, = ax2.plot(monthly_kpi["order_month"], monthly_kpi["late_rate(%)"], color=colors[2], linewidth=2, marker="o", label="지연율(%)")
ax2.set_ylabel("지연율(%)")
ax.set_title("월별 평균 리드타임 & 지연율\n(막대=평균 리드타임(일), 선=지연율(%))")
# 범례(두 축 핸들 통합)
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax.legend(h1+h2, l1+l2, loc="upper right", frameon=False)
plt.tight_layout(); plt.show()

# (E) 지연구간별 리뷰: 막대(저평점 비율) + 선(평균점수) — 의미를 제목에 명시
x = delay_bins_kpi["delay_bucket"]
plt.figure(figsize=(8,5))
ax = plt.gca()
bar = ax.bar(x, delay_bins_kpi["low_rate(%)"], color=colors[:len(x)], label="저평점 비율(≤3) [%]")
ax.set_ylabel("저평점 비율(≤3) [%]")
ax2 = ax.twinx()
line, = ax2.plot(x, delay_bins_kpi["review_mean"], color="#333333", marker="o", linewidth=2, label="평균 리뷰 점수")
ax2.set_ylabel("평균 리뷰 점수")
ax.set_title("지연구간별 리뷰 반응\n(막대=저평점 비율, 선=평균 리뷰 점수)")
# 범례(두 축 통합)
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax.legend(h1+h2, l1+l2, loc="upper left", frameon=False)
plt.tight_layout(); plt.show()

# ---------- 6) 간단 결론(프린트) ----------
def _line(k,v): print(f"- {k}: {v}")

print("\n=== 핵심 요약 ===")
_line("전체 평균 리드타임(일)", round(o["lead_days"].mean(),2))
_line("P90(일)", round(o["lead_days"].quantile(0.90),2))
_line("전체 지연율(%)", round(o["is_late"].mean()*100,2))
if pd.api.types.is_numeric_dtype(o["review_score"]):
    corr = pd.Series(o["delay_signed"]).corr(pd.Series(o["review_score"]))
    _line("지연(서명)↔리뷰 상관계수", round(corr,3))
_line("지연율 높은 주 TOP3", ", ".join(state_kpi.sort_values("late_rate(%)", ascending=False).head(3)["customer_state"].tolist()))
_line("물량 많은 주 TOP3", ", ".join(state_kpi.sort_values("orders", ascending=False).head(3)["customer_state"].tolist()))
```

해석 

버블맵: 우상단·큰 버블(리드타임↑, 지연율↑, 물량↑)이 개선 1순위 지역입니다.

월별 추이가 개선세면, 해당 기간의 물류 정책/프로모션 효과를 확인하세요.

지연-리뷰에서 >7일 지연의 저평점 비율이 급증하면, 선제 보상/알림 자동화가 ROI가 큽니다.

**lead_p90 = 총 리드타임(lead_days)의 ‘90번째 분위값’**이에요.
즉, 해당 주(state)의 주문 중 90%가 이 시간 이내에 고객에게 도착하고, 나머지 10%는 더 오래 걸린다는 뜻입니다. 평균보다 꼬리(지연) 영향을 덜 받으면서 상대적 최악 근처의 SLA 지표로 씁니다.

lead_median(중앙값)보다 lead_p90이 훨씬 크면 꼬리 지연이 큼 → 병목 구간 개선 필요

SLA 설정 시 “90% 주문은 X일 내” 같은 문구에 직접 사용 가능합니다.

지연구간별 리뷰 컬럼 설명

delay_bucket

지연 구간 라벨입니다. 생성 규칙은 배송일 − 약속일(estimated) 에 따라

정시/조기 : 지연 ≤ 0일(약속일보다 같거나 빨리 도착)

1-3일 지연 : 0 < 지연 ≤ 3

4-7일 지연 : 3 < 지연 ≤ 7

>7일 지연 : 지연 > 7

orders

해당 지연 구간에 속한 주문 수(고유 order_id 개수) 입니다. 분모가 커질수록 해당 구간의 사례가 많다는 뜻.

review_mean

해당 구간의 평균 리뷰 점수(1~5).

그룹 집계의 평균은 결측(미작성) 리뷰를 자동으로 제외하고 계산됩니다.

low_rate / low_rate(%)

저평점(≤3점) 비율입니다. 코드에서 low_rate = (review_score <= 3).mean()으로 계산했고,

low_rate(%)는 이를 백분율(×100, 소수점 반올림)로 표시한 값입니다.
